```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("RuralvsUrban2018.csv")
df
```

```python
df1= df[:30]

df1.set_index('Entity',inplace=True)
ax1 = df1[['RuralElectricity', 'UrbanElectricity']].plot(kind='bar', figsize=(10, 6))
ax1.set_title('Access to Electricity in Rural and Urban Areas(2018)')

plt.xticks(rotation=90)
plt.show()
```

```python
df2= df[30:60]

df2.set_index('Entity',inplace=True)
ax2 = df2[['RuralElectricity', 'UrbanElectricity']].plot(kind='bar', figsize=(10, 6))
ax2.set_title('Access to Electricity in Rural and Urban Areas(2018)')

plt.xticks(rotation=90)
plt.show()
```
