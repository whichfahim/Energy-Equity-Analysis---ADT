```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('datasets/data.csv')
```

```python
df.head()
```

```python
df2 = pd.read_csv('datasets/DEMO_DS.csv')
```

```python
display(df2)
```

```python
df2.rename(columns={'Time': 'Year'}, inplace=True)
df2.head()
```

```python
df2 = df2.drop('TIME', axis=1)
```

```python
df2.rename(columns={'LOCATION': 'code'}, inplace=True)
df2.head()
```

```python
df.rename(columns={'Code':'code'}, inplace=True)
```

```python
df.head()
```

```python
df2.head()
```

```python
result = pd.merge(df, df2, on=["code", "Year"], how="left")
result.head()
```

```python
display(result)
```

```python
result = result.drop('Country', axis=1)
result.head()
```

```python
result = result.drop('DEMO_IND', axis=1)
result.head()
```

```python
display(result)
```

```python
result = result[result.Year > 2015]
```

```python
display(result)
```

```python
result=pd.pivot_table(result,index=['code','Year','Access to electricity (% of population)'],columns='Indicator',values='Value',aggfunc=len,fill_value=0)
```

```python
display(result)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named df
# Replace 'Dependent_Column' and 'Independent_Column' with the actual column names in your DataFrame

# Plotting a scatter plot
plt.scatter(result['GDP (current US$)'], result['Access to electricity (% of population)'])

# Adding labels and title
plt.xlabel('GDP (current US$)')
plt.ylabel('Access to electricity (% of population)')
plt.title('Scatter Plot: GDP (current US$) vs Access to electricity (% of population)')

# Display the plot
plt.show()
```
